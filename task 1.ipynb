## Step 1 : Identify and handle missing values

```python
df.isnull()
```

```python
df.isnull().sum()
```

### Checked for missing or null values using df.isnull().sum(). No missing or null values rows were found in the dataset

## Step 2 : Remove duplicate rows

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.head()
```

### Checked for duplicate records using df.duplicated().sum(). No duplicate rows were found in the dataset.

## Step 3 : Standardize text values

```python
df["gender"] = df["gender"].str.lower()
```

```python
df["gender"].unique()
```

### Standardized text values in the gender column by converting all entries to lowercase for consistency.

## Step 4 : Convert date formats to a consistent type

```python
df.columns
```

### Checked the dataset for date columns. No date fields were available, so date format standardization was not applicable.

## Step 5 : Rename column headers to be clean and uniform

```python
df.columns
```

```python
df.columns = [
    'customer_id',
    'gender',
    'age',
    'annual_income',
    'spending_score'
]
```

```python
df.columns
```

### Renamed column headers to a standardized format by converting them to lowercase and replacing spaces/special characters with underscores.

```python
df.head()
```

## Step 6 : Check and fix data types 

```python
df.dtypes
```

### Verified data types for all columns. All variables were already stored in appropriate formats, so no data type conversions were necessary.
